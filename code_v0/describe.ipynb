```python
import pandas as pd
import geopandas as gpd
```

```python
def null_desc(df_col):
    
    # Describe Columns to get a better idea of what's going on
    print(df_col.describe())
    
    # Count Sum of nulls
    nl = df_col.isna().sum()
    
    # Percentage null
    nl_perc = nl/df_col.shape[0]
    
    
    print('Percentage of Null Values:', nl_perc, '%')
    
```

```python
def describe_df(file, shp=False, list_of_columns=[]):
    df = None
    
    # Reads df depending on if it's a shapefile or csv
    if shp == True:
        df = gpd.read_file(file)
    else:
        df = pd.read_csv(file)
    
    # Print the whole file or not specified by the column list
    if len(list_of_columns) < 1:
        
        for col in df.columns:
             # Geometry takes awhile to average so avoiding it is a good idea
            if col == 'geometry':
                continue
            print('Starting Column:', col)
            null_desc(df[col])
            
    else:
        for col in list_of_columns:
            
            # Geometry takes awhile to average so avoiding it is a good idea
            if col == 'geometry':
                continue
            print('\nStarting Column:', col)
            null_desc(df[col])
        
       
```

```python
file2 = '../data/streets/cbb130b3-3f8e-4680-ac2c-ab1f84527d762020329-1-1mrz2ha.zxri.shp'

describe_df(file2, shp=True)
```
